```python
df_raw = df_raw.dropna()
df_raw.isnull().sum() 
```

```python
filter_col = [col for col in df_raw if col.startswith('cg')]
print(len(filter_col))
```

This is too much to apply EDA. Therefore, I will employ PCA for dimensionality reduction.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

df_for_PCA = df_raw[filter_col]

# You must normalize the data before applying the fit method
df_normalized=(df_for_PCA - df_for_PCA.mean()) / df_for_PCA.std()
pca = PCA(n_components=41)
pca.fit(df_normalized)

# Reformat and view results
loadings = pd.DataFrame(pca.components_.T,
columns=['PC%s' % _ for _ in range(41)],
index=df_for_PCA.columns)
print(loadings)

plt.plot(pca.explained_variance_ratio_)
plt.ylabel('Explained Variance')
plt.xlabel('Components')
plt.show()
```

Since min(n_samples, n_features)=41, we can assign output length as 41. For mor explanation: https://builtin.com/data-science/step-step-explanation-principal-component-analysis

```python
loadings = pd.DataFrame(pca.components_.T,
columns=['PC%s' % _ for _ in range(41)])

df = pd.concat([loadings, df_raw["Age"]], axis=1, join="inner")
df.head(-2)
```

Let's see the correlations for each column including Age.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Find correlations between columns
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

Let's train a tree based model (from my experience, tree based models are quite effective on tabular data). Therefore, second approach will be Random Forest. For more: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

```python
from sklearn.model_selection import train_test_split

X = df[['PC%s' % _ for _ in range(41)]]
y = df[["Age"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=48)
```

Dataset is split for training and test.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

# Setting random_state so that we can get same results (consistency in terms of producing same results)
regr = RandomForestRegressor(max_depth=10, random_state=0)
regr.fit(X_train, y_train)

score = regr.score(X_train, y_train)
print("R-squared:", score) 
```

I picked the evaluation metric the coefficient of determination for training, because it is the proportion of the variation in the dependent variable that is predictable from the independent variable(s). Source: https://www.geeksforgeeks.org/ml-r-squared-in-regression-analysis/

```python
from sklearn.metrics import mean_squared_error

y_pred = regr.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("MSE: ", mse)
print("RMSE: ", mse*(1/2.0)) 
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, linewidth=1, label="original")
plt.plot(x_ax, y_pred, linewidth=1.1, label="predicted")
plt.title("y-test and y-predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```

Results are not good at all.
